```python
import os
import glob

folder_path = 'C:/Users/GnTRnD/Desktop/gen_ai/zero_city/zero_city_track'  # Replace with the actual folder path

# Get a list of all text files in the folder
text_files = glob.glob(os.path.join(folder_path, '*.txt'))

# Iterate over the text files
for text_file in text_files:
    image_file = text_file[:-4] + '.jpg'  # Assuming image files have .jpg extension, change it accordingly
    
    # Check if the corresponding image file exists
    if not os.path.exists(image_file):
        # Remove the text file
        os.remove(text_file)
        print(f"Removed {text_file}")

print("Done removing text files without a matching image.")
```
