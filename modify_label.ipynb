```python
import os
from PIL import Image

def convert(size, box):
    dw = size[0]
    dh = size[1]
    x_min = box[0]
    y_min = box[1]
    x = (x_min + box[2] / 2) / dw
    y = (y_min + box[3] / 2) / dh
    w = box[2] / dw
    h = box[3] / dh
    return (x, y, w, h)

# Provide the necessary information
image_folder = "C:/Users/GnTRnD/Desktop/gen_ai/zero_city/zero_city_track"
label_folder = os.path.join(image_folder, "labels")

# Get a list of image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.endswith(".jpg")]

# Process each image and its corresponding label file
for image_filename in image_files:
    # Get the label file name corresponding to the image file
    label_filename = image_filename.replace(".jpg", ".txt")
    image_path = os.path.join(image_folder, image_filename)
    label_path = os.path.join(label_folder, label_filename)

    # Read the image file to get its width and height
    im = Image.open(image_path)
    width = int(im.size[0])
    height = int(im.size[1])

    # Read the label file and modify the bounding box values
    with open(label_path, "r") as file:
        lines = file.readlines()

    modified_lines = []
    for line in lines:
        values = line.split(" ")
        class_id = int(values[0])
        xmin = float(values[1])
        xmax = float(values[2])
        ymin = float(values[3])
        ymax = float(values[4])

        # Define the bounding box coordinates
        bounding_box = (xmin, xmax, ymin, ymax)

        # Convert the bounding box to YOLO format
        yolo_box = convert((width, height), bounding_box)

        # Create the modified line with the converted YOLO format bounding box
        modified_line = f"{class_id} {yolo_box[0]} {yolo_box[1]} {yolo_box[2]} {yolo_box[3]}\n"
        modified_lines.append(modified_line)

    # Write the modified label file
    with open(label_path, "w") as file:
        file.writelines(modified_lines)

    print(f"Converted bounding box values for {image_filename}")
```

```python
import os

label_folder = "C:/Users/GnTRnD/Desktop/gen_ai/zero_city/labels_yolo"

# Get a list of label files in the folder
label_files = [f for f in os.listdir(label_folder) if f.endswith(".txt")]

# Define the class_id mapping
class_id_mapping = {
    39: 0,
    56: 1,
    41: 2,
    57: 3,
    40: 4,
    42: 5,
    43: 6,
    58: 7,
    50: 8,
    51: 9,
    52: 10,
    125: 11,
    44: 12,
    45: 13,
    46: 14,
    47: 15,
    48: 16,
    53: 17
}

# Process each label file
for label_filename in label_files:
    label_path = os.path.join(label_folder, label_filename)

    # Read the label file and modify the class_id values
    with open(label_path, "r") as file:
        lines = file.readlines()

    modified_lines = []
    for line in lines:
        values = line.split(" ")
        class_id = int(values[0])

        # Replace the class_id value if it exists in the mapping dictionary
        if class_id in class_id_mapping:
            class_id = class_id_mapping[class_id]

        # Create the modified line with the new class_id value
        modified_line = f"{class_id} {' '.join(values[1:])}"
        modified_lines.append(modified_line)

    # Write the modified label file
    with open(label_path, "w") as file:
        file.writelines(modified_lines)

    print(f"Replaced class_id values in {label_filename}")
```
